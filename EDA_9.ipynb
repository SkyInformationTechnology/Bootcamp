```python
out = [ ]
def Zscore_outlier(df):
  m = np.mean(df)
  sd = np.std(df)
  for i in df :
    Z = (i-m)/sd
    if np.abs(Z)>3:
      out.append(i)
  print("outliers:",out)
  Zscore_outlier(df['fixed acidity'])
```
